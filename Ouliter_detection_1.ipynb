# Percentile

```python
import pandas as pd
```

```python
df=pd.read_csv('heights.csv')
```

```python
df.head()
```

```python
maxthres=df.height.quantile(.95)
```

```python
df[df['height']>maxthres]
```

```python
minthres=df.height.quantile(.05)
```

```python
df[df['height']<minthres]
```

```python
df[(df['height']<maxthres) & (df['height']>minthres)]
```

```python
df2=pd.read_csv('bhp.csv')
```

```python
df2.describe()
```

```python
min_threshold,max_threshold=df2.price_per_sqft.quantile([0.001,0.999])
```

```python
min_threshold,max_threshold
```

```python
df2[(df2.price_per_sqft<min_threshold)]
```

```python
df2[(df2.price_per_sqft>max_threshold)]
```

```python
hp=df2[(df2.price_per_sqft<max_threshold) & (df2.price_per_sqft>min_threshold)]
```
